# Membaca file csv dan menjalankannya

Di tugas ini saya akan menghilangkan nilai di atribut yang masih bernilai null / Nan, supaya saat modelling data,
hasilnya akan lebih baik dan akurat

```python
data = pd.read_csv("water_potability.csv") 
```

```python
data.head(20)
```

di tahap ini saya memisahkan antara atribut dan label, supaya proses data preparation berjalan dengan baik di tahap atribut

```python
atribut = data.iloc[:,:-1].values
label = data.iloc[:,-1].values
print(atribut)
```

Setelah memisahkan atribut dan label, kemudian saya menggunakan library sklearn.impute dengan mengimport SimpleImputer yang 
berfungsi mengisi nilai di atribut yang masih bernilai nan / null / tdk terdefinisi. Untuk perbedaan bisa dilihat di tampilan
print(atribut) yang pertama masih ada nilai nan, kemudian setelah saya gunakan sklearn.impute dengan mengimport SimpleImputer
nilai nan menjadi hilang

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(atribut[:,:-1])
atribut[:,:-1] = imp_mean.transform(atribut[:,:-1])
print(atribut)
```
