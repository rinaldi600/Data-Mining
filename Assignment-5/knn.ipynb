Dibawah ini merupakan beberapa library yang saya gunakan yaitu pandas, numpy dan matplotib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 9999)
pd.set_option('display.max_columns', 9999)
pd.set_option('display.width', 9999)
```

# Menjalakan dataset dengan read_csv fungsi dari library pandas

```python
data = pd.read_csv("healthcare-dataset-stroke-data.csv") 
```

```python
data.head(300)
```

Dibawah ini saya encoding di beberapa atribut yang masih menggunakan tipe categorical, karena jika tidak saya ubah ke numerik nantinya akan terjadi error saat menjalakan algoritma gaussian naive bayes ini

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
data['gender'] = labelEncoder.fit_transform(data['gender'])
data['ever_married'] = labelEncoder.fit_transform(data['ever_married'])
data['work_type'] = labelEncoder.fit_transform(data['work_type'])
data['Residence_type'] = labelEncoder.fit_transform(data['Residence_type'])
data['smoking_status'] = labelEncoder.fit_transform(data['smoking_status'])
data
```

```python
atributData = data.iloc[:,:-1].values
labelData = data.iloc[:,-1].values
```

Dibawah ini saya menghilangkan atribut missing value di beberapa atribut guna penyebaran data merata

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mean.fit(atributData[:,:-1])
atributData[:,:-1] = imp_mean.transform(atributData[:,:-1])
```

Dibawah ini saya membagi antara data training dengan data testing

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X_train, X_test, y_train, y_test = train_test_split(atributData, labelData, test_size=0.33, random_state=0)
```

Dibawah ini saya melakukan scalling data guna memperpendek jarak antar nilai

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

Dibawah ini saya menjalankan gaussian dari sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

Dibawah ini saya melakukan prediksi dari daatset yang sudah saya jalankan dengan algoritma gaussian naive bayes hasilnya mempunyai akurasi sebesar 95% dari skala 100%

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

Dibawah ini saya menghitung juga matrik confusion errornya

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
