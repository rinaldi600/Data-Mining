```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import dataframe_image as dfi
```

```python
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('Tweets.csv')
```

```python
data.columns = data.columns.str.lower()
```

# Case Folding

Case folding merupakan proses untuk mengubah kalimat atau menjadi huruf kecil

```python
data = data.apply(lambda x: x.astype(str).str.lower())
```

```python
data
```

# Data Cleaning

Data cleansing adalah kegiatan menganalisa kualitas data dengan cara memodifikasi, mengubah, atau 
menghapus data-data yang dianggap tidak perlu, tidak lengkap, data tidak akurat, atau memiliki format
data atau file yang salah dalam basis data guna menghasilkan data berkualitas tinggi

```python
def removeHastagUserEtc(text) :
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",text).split())
```

```python
data['text'] = data['text'].apply(removeHastagUserEtc)
data['username'] = data['username'].apply(removeHastagUserEtc)
```

```python
data['username'] = data['username'].apply(lambda x: x.lower().replace(' ', ''))
```

```python
data
```

# Stopword Removal

Stopword removal disebut juga filtering, adalah tahap pemilihan kata-kata penting dari hasil token, yaitu 
kata-kata apa saja yang akan digunakan untuk mewakili dokumen.

```python
def stopWordRemovalTextIndonesia(text) :
    textTweet = text.split()
    stopWord = stopwords.words('english')
    resultWord = [word for word in textTweet if word not in stopWord]
    return ' '.join(resultWord)
```

```python
data['text'] = data['text'].apply(stopWordRemovalTextIndonesia)
```

```python
data
```

# Labelling

labelling merupakan teknik untuk menentukan apakah tweets ini bernilai positif atau negatif

```python
sentiment = []
for x in data['text'] :
    analysis = TextBlob(x)
    if analysis.sentiment.polarity >= 0.05 :
        sentiment.append("Positive")
    else :
        sentiment.append("Negative")
    print(str(analysis.sentiment.polarity) , " : ", x)
```

```python
data['sentiment'] = sentiment
```

```python
data
```

```python
data = data.drop('username', axis=1)
```

# Naive Bayes Classifier

```python
x = data['text']
```

```python
y = data['sentiment']
```

```python
x, x_test, y, y_test = train_test_split(x,y, stratify=y, test_size=0.25, random_state=1)
```

```python
counter = CountVectorizer()
counter.fit(x, y)
```

```python
train_counts = counter.transform(x)
test_counts = counter.transform(x_test)
```

```python
classifier = MultinomialNB()
classifier.fit(train_counts, y)

prediction = classifier.predict(test_counts)
```

```python
print(accuracy_score(y_test, prediction))
```

```python
confusion_matrix(y_test, prediction)
```

```python
classifier.predict(counter.transform(['stop bullying for child']))
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, prediction))
```

```python
disp.plot()
plt.show()
```

```python
print(classification_report(y_test, prediction))
```
